```python
import pandas as pd
```

```python
training_data = pd.read_csv("/content/sample_data/california_housing_train.csv")
```

```python
testing_data = pd.read_csv("/content/sample_data/california_housing_test.csv")
```

```python
training_data
```

```python
training_data = (training_data - training_data.mean())/training_data.std()
```

```python
training_data
```

```python
testing_data = (testing_data - testing_data.mean())/testing_data.std()
```

```python
testing_data
```

```python
X_dash = training_data.iloc[:,0:8]
X_dash
```

```python
Y_dash = training_data["median_house_value"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_cv,Y_train,Y_cv = train_test_split(X_dash,Y_dash,test_size=0.2, random_state=42)
```

```python
import numpy as np
```

```python
X_train = np.array(X_train)
X_cv= np.array(X_cv)
Y_train= np.array(Y_train)
Y_cv= np.array(Y_cv)
```

```python
from keras.layers import InputLayer,Dense,BatchNormalization
from keras.models import Sequential
```

```python
from keras.regularizers import l2
from keras.optimizers import Adam
```

```python
def build_network():

  network = Sequential()
  
  network.add(InputLayer(input_shape=(X_train.shape[1],),))
 
  #kernel_initializer = connection matrix of theta's
  network.add(Dense(units=X_train.shape[1],activation="relu",kernel_initializer="glorot_normal",kernel_regularizer=l2()))
  
  network.add(BatchNormalization())
  
  network.add(Dense(units=X_train.shape[1],activation="relu",kernel_initializer="glorot_normal",kernel_regularizer=l2()))
  
  network.add(BatchNormalization())
  
  network.add(Dense(units=1,kernel_initializer="glorot_normal",kernel_regularizer="l2"))

  return network
```

```python
network = build_network()
```

```python
network.compile(optimizer="adam",loss="mse")
```

```python
performance_history = network.fit(x=X_train,y=Y_train,batch_size=136,epochs=100 ,validation_data=(X_cv,Y_cv))
```

```python
performance_history.history
```

```python
training_loss = performance_history.history['loss']
cv_loss = performance_history.history['val_loss']
```

```python
epochs = range(1,len(training_loss)+1)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(epochs,training_loss,'ro',label="Training Loss")
plt.plot(epochs,cv_loss,'b',label="Cross validation Loss")
plt.title("training and cross vlidation loss")
plt.xlabel("epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
avg_training_loss = sum(training_loss)/100
avg_training_loss
```

```python
avg_cv_loss = sum(cv_loss)/100
avg_cv_loss
```
